```python

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select 

import pandas as pd
import numpy as np

import time
```

# 2023 - 2012

```python
scores = []
countries = []
years = []

for y in range(2012,2024):
    # create webdriver object
    driver = webdriver.Chrome(r'C:\Users\panda\Downloads\chromedriver-win32\chromedriver')

    # get to the website
    driver.get(f"https://www.transparency.org/en/cpi/{y}")

    time.sleep(5)

    score = driver.find_elements(By.CSS_SELECTOR,"span.font-bold.inline-block.mr-4")
    for s in score:
        scores.append(s.text)
    # print(len(scores))
    country = driver.find_elements(By.CSS_SELECTOR,"span.flex-1.truncate.pr-2")
    for c in country:
        countries.append(c.text)
        years.append(y)
    print(y, len(scores), len(years))
    driver.close()
```

```python
dt = pd.DataFrame({'country': countries,
                    'year': years,
                    'score': scores})
first_last = dt.sort_values(['country','year']).groupby(dt['country'])['score'] \
        .agg(['first', 'last']).reset_index()
first_last = first_last.rename(columns={"first": "Earliest", "last": "MostRecent"})
dt = dt.pivot_table(index = 'country',
                    columns='year',
                    values = 'score').reset_index()
dt = dt.merge(first_last, on = 'country', how = 'left')
country = pd.read_excel('Country.xlsx')
dt = country.merge(dt, on = 'country', how = 'left')
dt = dt.drop(columns=['country'])
# dt.to_csv('CorruptionCPINew.csv',index = False)
dt
```
